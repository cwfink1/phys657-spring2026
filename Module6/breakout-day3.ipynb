## Discussion

**Why can gradient boosting overfit with too many trees, while random forests cannot?**

Think about:
- In a random forest, each new tree is trained independently on a fresh bootstrap sample. What happens when you average one more independent estimate?
- In gradient boosting, each new tree is trained to correct the *residual errors* of the current ensemble. What happens when the remaining residuals are mostly noise?
- How does this relate to the difference between averaging more measurements (reduces noise) vs fitting a higher-order polynomial (eventually fits noise)?
